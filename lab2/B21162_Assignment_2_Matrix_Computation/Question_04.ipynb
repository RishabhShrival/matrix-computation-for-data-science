```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def leibniz(n):
  pi = 0
  for i in range(n):
    pi = pi+((-1)**i)*(1/(2*i +1))

  return pi *4
```

```python
def diff(n):
  d = abs(math.pi - leibniz(n))
  return d
```

```python
X = []
Y = []
for i in range(100):
  X.append(i)
  Y.append(diff(i))
```

```python
plt.plot(X,Y)
plt.show()
```
